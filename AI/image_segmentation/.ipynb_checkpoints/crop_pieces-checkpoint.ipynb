```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('./sample_blacklabel.png')
origin = img.copy()

imgray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret, th = cv2.threshold(imgray, 127, 255, cv2.THRESH_BINARY_INV)

contours, hr = cv2.findContours(th, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)[-2:]
contr = contours[0]

x, y, w, h = cv2.boundingRect(contr)
cv2.rectangle(img, (x, y), (x+w, y+h), (0,0,0), 3)

rect = cv2.minAreaRect(contr)
box = cv2.boxPoints(rect)
box = np.int0(box)
cv2.drawContours(img, [box], -1, (0, 255, 0), 2)

[vx, vy, x, y] = cv2.fitLine(contr, cv2.DIST_L2, 0, 0.01, 0, 0.01)
cols, rows = img.shape[:2]
cv2.line(img, (0, int(0-x*(vy/vx) + y)), (cols-1, int((cols-x)*(vy/vx) + y)), (0,0,255), 2)

plt.figure(figsize=(10, 6))
imgs = {'origin':origin, 'Bound Fit shapes':img}

for i, (k, v) in enumerate(imgs.items()):
    plt.subplot(1, 2, i+1)
    plt.title(k)
    plt.imshow(v[:,:,(2,1,0)])
    plt.xticks([]), plt.yticks([])

plt.show()
```

```python
import selectivesearch
import cv2
import matplotlib.pyplot as plt
import os

img = cv2.imread('./sample_blacklabel.png')
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
print("image shape",img.shape)

plt.figure(figsize=(8,8))
plt.imshow(img_rgb)

plt.show()
```

```python
import selectivesearch

#scale 크다 -> 큰 object 위주로 찾아라, min_size -> 적어도 min_size 이상의 object 찾아라
#regions -> object가 있을만한 공간
_, regions = selectivesearch.selective_search(img_rgb, scale=100, min_size=4)

print(type(regions), len(regions))


cand_rects = [cand['rect'] for cand in regions if cand['size'] > 10000] # rects만 가져와서 저장

green_rgb = (125,255,51)
img_rgb_copy = img_rgb.copy() # 그림 카피 뜨기

for rect in cand_rects:
  left=rect[0]
  top=rect[1]

  right = left + rect[2]
  bottom = top + rect[3]

  img_rgb_copy = cv2.rectangle(img_rgb_copy, (left,top),(right,bottom),color=green_rgb, thickness=2) 
  #사각형 좌표의 왼쪽 상단, 오른쪽 하단 입력

plt.figure(figsize=(8,8))
plt.imshow(img_rgb_copy)
plt.show()
```

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
import glob
import matplotlib.image as mpimg

def split_and_crop(image_path_list):
    # 크롭된 영역을 저장할 리스트
    cropped_areas = []

    # 문제 최소 크기
    min_width, min_height = 100, 100
    
    for image_path in image_path_list:
        #원본 이미지
        image = cv2.imread(image_path)

        plt.title('INPUT IMAGE')
        plt.imshow(image)
        plt.xticks([]), plt.yticks([])
        plt.show()

        image_l = image[60:image.shape[0]-60, :image.shape[1]//2 - 5]
        image_r = image[60:image.shape[0]-60, image.shape[1]//2 + 5:]

        for image in (image_l, image_r):
            #흑백 이미지
            imgray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) #원본 이미지와 흑백이미지가 따로 존재
        
        	#이미지 임계화
            blur = cv2.GaussianBlur(imgray, ksize=(3,3), sigmaX=0)
            thresh = cv2.threshold(blur, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)[1]
        
        
            # Morph operations
            edge = cv2.Canny(imgray, 100, 200)
        
        	#RECT Shape, Dilation이용 
            kernel = cv2.getStructuringElement(cv2.MORPH_RECT,(20, 20))
            dilated = cv2.morphologyEx(edge, cv2.MORPH_CLOSE, kernel)
        
            kernel = cv2.getStructuringElement(cv2.MORPH_RECT,(15,15))
            dilated = cv2.morphologyEx(edge, cv2.MORPH_DILATE, kernel)
        
            # kernel = cv2.getStructuringElement(cv2.MORPH_RECT,(5, 15))
            # dilated = cv2.morphologyEx(edge, cv2.MORPH_CLOSE, kernel)
        
        	#출력 
            # plt.title('dilated 1')
            # plt.imshow(dilated)
            # plt.xticks([]), plt.yticks([])
        
        
            contours, hierarchy = cv2.findContours(dilated.copy(),cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
        
        
            
            
            for contour in contours:
                # 경계를 둘러싸는 최소 사각형 계산
                x, y, w, h = cv2.boundingRect(contour)
                
                if w >= min_width and h >= min_height:
                    # 원본 이미지에서 해당 영역 크롭
                    cropped = image[y:y+h, x:x+w]
                    
                    # 크롭된 영역을 리스트에 추가
                    cropped_areas.append(cropped)
            
         #    #원본 이미지 위에 윤곽 그리기
         #    contours_image = cv2.drawContours(image, contours, -1, (0,255,0), 1)
        
            
        	# #출력 
         #    plt.title('contours_image')
         #    plt.imshow(contours_image)
         #    plt.xticks([]), plt.yticks([])
        
         #    plt.show()

    return cropped_areas
    
    
#함수실행     
if __name__ == "__main__":

    # 특정 폴더 지정
    image_path = './'
    
    # 해당 폴더 내의 모든 .png 파일 검색
    image_path = glob.glob(f"{image_path}/*.png")
    
    problem_list = split_and_crop(image_path)
    
    # 결과 확인
    plt.figure(figsize=(10,12))
    for i, cropped in enumerate(problem_list):
        plt.subplot(3,2,i+1)
        plt.xticks([]), plt.yticks([])
        
        plt.title(f'Cropped {i+1}')
        plt.imshow(cropped)

    plt.show()


## Template Matching
```
